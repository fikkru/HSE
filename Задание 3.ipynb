Задание1
Переведитесодержимоефайлаpurchase_log.txtвсловарьpurchasesвида:
{'1840e0b9d4': 'Продукты', ...}
Пример работы программы при выводе первых двух элементов словаря
purchases:
1840e0b9d4 ‘Продукты‘
4e4f90fcfb ‘Электроника‘

```python
import json

purchases = {}

with open('purchase_log.txt', 'r', encoding='utf-8') as file:
    for line in file:
        line = line.strip()
        if line:
            try:
                data = json.loads(line)
                user_id = data.get('user_id')
                category = data.get('category')
                
                if user_id and category and user_id != 'user_id':
                    purchases[user_id] = category
            except json.JSONDecodeError:
                continue

print("Первые два элемента словаря purchases:")
count = 0
for user_id, category in purchases.items():
    if count < 2:
        print(f"{user_id} '{category}'")
        count += 1
    else:
        break
```

Задание 2
Для каждого user_id в файле visit_log.csv определите третий столбец с
категорией покупки. Если покупка была, сам файл visit_log.csv изменять не
надо.
Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были
покупки с указанием категории.
Учтите условия для данных:
● содержимое purchase_log.txt помещается в оперативную память
компьютера;
● содержимое visit_log.csv — нет, используйте только построчную
обработку этого файла.
Примеры работы программы при выводе первых трёх строк файла funnel.csv:
user_id,source,category
1840e0b9d4,other,Продукты
4e4f90fcfb,context,Электроника

```python
import json

purchases = {}

with open('purchase_log.txt', 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if line:
            try:
                data = json.loads(line)
                if data['user_id'] != 'user_id': 
                    purchases[data['user_id']] = data['category']
            except:
                pass

with open('visit_log.csv', 'r', encoding='utf-8') as visit_f, \
     open('funnel.csv', 'w', encoding='utf-8') as funnel_f:
    
    funnel_f.write('user_id,source,category\n')
    
    visit_f.readline()
    
    for line in visit_f:
        parts = line.strip().split('\t')  
        
        if len(parts) >= 2:
            user_id = parts[0]
            
            if user_id in purchases:
                source = parts[1]
                category = purchases[user_id]
                funnel_f.write(f'{user_id},{source},{category}\n')

print("Первые строки файла funnel.csv:")
with open('funnel.csv', 'r', encoding='utf-8') as f:
    count = 0
    for line in f:
        if count < 4:  
            print(line.strip())
            count += 1
        else:
            break
```
